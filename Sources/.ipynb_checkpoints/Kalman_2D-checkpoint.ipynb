```python
import numpy.matlib
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
# Initialize estimated state mean vector and variance matrix (position and velocity: x, y, vx, vy)
x = np.matrix([1,1,0,0]).T
P = np.matlib.eye(4, 4)

# Increase variance of velocity because it is unknown (but position in known)
P[2,2] = P[3,3] = 100 

# Initialize process noise variance matrix
Q = 1e-6 * np.matlib.eye(4, 4)

# Initialize transition matrix
dt = 1/30 # 30 FPS
F = np.matlib.eye(4, 4)
F[0,2] = F[1,3] = dt 

# Initialize measurement matrix
H = np.matlib.eye(2, 4)

# Initialize measurement noise variance matrix
R = 1e-3 * np.matlib.eye(2, 2)
```

```python
steps = 100
predictions = np.matlib.zeros((4, steps))
measurements = np.matlib.zeros((2, steps))
```

```python
# Simulate 2D signal change in time sequence
t = np.linspace(0,20,steps)
xline = np.sin(t)
yline = np.cos(t)

zx = xline + np.matrix(np.random.multivariate_normal([1,1],[[0.5,0],[0,0.5]])).T
zy = yline + np.matrix(np.random.multivariate_normal([1,1],[[0.5,0],[0,0.5]])).T

fig = plt.figure()
axes3d = Axes3D(fig)
axes3d.plot(xline,yline,t)
plt.xlabel('X',size = 30)
plt.ylabel('Y',size = 30)
axes3d.set_zlabel('time sequence',color = 'k',size=30)
axes3d.set_yticks([-1,0,1])
axes3d.set_yticklabels(['min',0,'max'])
```

```python
for t in range(0,steps):

    # Measurement
    zt = np.matrix(np.random.multivariate_normal([1,1],[[0.5,0],[0,0.5]])).T
    
    
    # Prediction
    xt = F * x
    Pt = F * P * F.T + Q
    
    # Update
    K = Pt * H.T * np.linalg.inv(H * Pt * H.T + R)
    x = xt + K * (zt - H*xt)
    P = Pt - K * H * Pt

    # Record
    predictions[:,t] = x
    measurements[:,t] = zt

```

```python
# query data
position_x = predictions[0,:].getA1()
position_y = predictions[1,:].getA1()
velocity_x = predictions[2,:].getA1()
velocity_y = predictions[3,:].getA1()
measurement_x = measurements[0,:].getA1()
measurement_y = measurements[1,:].getA1()

# plot
plt.rcParams['figure.figsize'] = (12, 8)
plt.figure()
plt.subplot(2,2,1)
plt.plot(measurement_x, label='measurements_x')
plt.plot(position_x, label='predictions_x')
plt.legend()
plt.subplot(2,2,2)
plt.plot(measurement_y, label='measurements_y')
plt.plot(position_y, label='predictions_y')
plt.legend()
plt.subplot(2,2,3)
plt.plot(velocity_x, label='velocity_x')
plt.legend()
plt.subplot(2,2,4)
plt.plot(velocity_y, label='velocity_y')
plt.legend()
plt.show()
```
